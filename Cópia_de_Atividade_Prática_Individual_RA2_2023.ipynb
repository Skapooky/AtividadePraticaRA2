```python
#Definindo precisão
mp.prec = 100
mp.dps = 4

#Definido f(x)
def g(x):
  return (x**4) - (5*x**3) + (10*x**2) - (10*x +5)

#Definindo segunda derivada
def derivada2(x):
  return 8*x**2 - 30*x + 20

#Criando vetor de x
xValues = np.linspace(-11, 11, 1000)

#Plotando Gráfico
fig = go.Figure()
fig.add_trace(go.Scatter(x=xValues, y=g(xValues), mode='lines'))
fig.show()

#Achar pontos críticos algébricos
x = sy.Symbol('x')
f_x = (x**mpf(4)) - (mpf(5)*x**mpf(3)) + (mpf(10)*x**mpf(2)) - (mpf(10)*x +mpf(5))

f_prime_x = sy.diff(f_x, x)
pontos_criticos = sy.solve(f_prime_x, x)

#print(derivada2(pontos_criticos[0])) é menor que 0

print(f"Primeiro ponto mínimo: {pontos_criticos[0]}")
```

# Questão 3 (vale 2 pontos)
Considerando a função $f(x,y) = \sin(xy) + \cos(x+y)$, no intervalo $0\leq x\leq 5 \space\space e \space\space 0\leq y\leq 5$ trace, usando o plotly o gráfico desta função e marque os pontos de máximo e mínimo neste intevalo.

Lembre-se o sympy é sempre uma opção e você pode usar qualquer técnica, algoritmo, regra, ou tabela, que tenha sido usada na matéria. E que precisa de apenas uma casa decimal de precisão. Uma dica importante para a solução deste problema é a função **np.unravel_index**


```python
def f(x, y):
    return np.sin(x*y) + np.cos(x+y)

x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig1 = go.Figure(data=[go.Surface(x=X, y=Y, z=Z)])
fig1.update_layout(title="Gráfico de f(x,y) = sin(xy) + cos(x+y)", autosize=False, width=600, height=600)
fig1.show()
```
